```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# -------------------------------
# 1) LOAD DATA
# -------------------------------
df = pd.read_csv("dataset1/kidney_disease.csv")

# Replace "?" or NaN values
df.replace("?", np.nan, inplace=True)

# -------------------------------
# 2) CLEANING
# -------------------------------
# Convert numeric columns to proper dtype
numeric_cols = ["age","bp","sg","al","su","bgr","bu","sc","sod","pot","hemo","pcv","wc","rc"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Fill numeric missing values with median
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Fill categorical missing values with mode
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna(df[col].mode()[0])

# -------------------------------
# 3) ENCODING
# -------------------------------
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = le.fit_transform(df[col])

# -------------------------------
# 4) SPLIT
# -------------------------------
X = df.drop("classification", axis=1)   # features
y = df["classification"]                # target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# -------------------------------
# 5) TRAIN MODEL (High Accuracy)
# -------------------------------
rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=10,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42
)

rf.fit(X_train, y_train)

# -------------------------------
# 6) EVALUATE
# -------------------------------
y_pred = rf.predict(X_test)

acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print("Accuracy:", acc)
print("\nConfusion Matrix:\n", cm)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# --- Convert to binary labels, drop the tiny noisy class, retrain and evaluate ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, roc_curve, auc, classification_report
)
import seaborn as sns

# assume df is already loaded
print("Before mapping:", df['classification'].value_counts())

# 1) Map labels: 0 -> 0 (normal), 2 -> 1 (decline)
df = df.copy()
df = df[df['classification'] != 1]          # drop the noisy class with 2 samples
df['classification'] = df['classification'].replace({0: 0, 2: 1})

print("After mapping:", df['classification'].value_counts())

# 2) Ensure numeric columns are numeric (same columns list as before)
numeric_cols = ["age","bp","sg","al","su","bgr","bu","sc","sod","pot","hemo","pcv","wc","rc"]
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Simple imputation for numeric columns (median)
num_present = [c for c in numeric_cols if c in df.columns]
df[num_present] = df[num_present].fillna(df[num_present].median())

# Fill categorical missing with mode
for c in df.columns:
    if df[c].dtype == "object":
        if not df[c].mode().empty:
            df[c] = df[c].fillna(df[c].mode().iloc[0])
        else:
            df[c] = df[c].fillna("unknown")

# 3) Encode any remaining object columns
le = LabelEncoder()
for c in df.columns:
    if df[c].dtype == "object":
        df[c] = le.fit_transform(df[c])

# 4) Prepare X, y
X = df.drop("classification", axis=1)
y = df["classification"]

# 5) Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# 6) Scale numeric features for LogisticRegression
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

# 7) Train RandomForest
rf = RandomForestClassifier(n_estimators=300, max_depth=10, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:,1]

# 8) Train Logistic Regression (for comparison)
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train_s, y_train)
y_pred_lr = lr.predict(X_test_s)
y_prob_lr = lr.predict_proba(X_test_s)[:,1]

# 9) Evaluation function
def eval_print(y_true, y_pred, y_prob, model_name="Model"):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    roc_auc = auc(fpr, tpr)
    print(f"\n--- {model_name} ---")
    print(f"Accuracy: {acc:.4f}  Precision: {prec:.4f}  Recall: {rec:.4f}  F1: {f1:.4f}  AUC: {roc_auc:.4f}")
    print(classification_report(y_true, y_pred, zero_division=0))
    return fpr, tpr, roc_auc

fpr_rf, tpr_rf, auc_rf = eval_print(y_test, y_pred_rf, y_prob_rf, "RandomForest")
fpr_lr, tpr_lr, auc_lr = eval_print(y_test, y_pred_lr, y_prob_lr, "LogisticRegression")

# 10) Plot confusion matrix (RF) and ROC curves
fig, axes = plt.subplots(1,2, figsize=(12,5))

# Confusion matrix RF
cm = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title("RandomForest Confusion Matrix")
axes[0].set_xlabel("Predicted")
axes[0].set_ylabel("Actual")
axes[0].xaxis.set_ticklabels(['normal','decline'])
axes[0].yaxis.set_ticklabels(['normal','decline'])

# ROC curve comparison
axes[1].plot(fpr_rf, tpr_rf, label=f'RF AUC={auc_rf:.3f}')
axes[1].plot(fpr_lr, tpr_lr, label=f'LR AUC={auc_lr:.3f}')
axes[1].plot([0,1],[0,1],'--', color='gray')
axes[1].set_xlabel('False Positive Rate')
axes[1].set_ylabel('True Positive Rate')
axes[1].set_title('ROC Curves')
axes[1].legend()

plt.tight_layout()
plt.show()

# 11) Feature importance (RandomForest)
import numpy as np
feat_imp = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nTop 10 feature importances (RF):")
print(feat_imp.head(10))

plt.figure(figsize=(6,4))
feat_imp.head(10).plot(kind='barh')
plt.gca().invert_yaxis()
plt.title("RandomForest Top-10 Feature Importance")
plt.show()

# 12) Save cleaned CSV (optional)
df.to_csv("kidney_data_cleaned_binary.csv", index=False)
print("Saved cleaned binary CSV: kidney_data_cleaned_binary.csv")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# --- Diagnostic: show unique labels in full, train, test (if split exists) ---
print("Full label distribution:\n", y.value_counts())

# If current X_train/y_train exist, show them; otherwise we will resplit
try:
    print("Train labels unique:", np.unique(y_train), "counts:", np.unique(y_train, return_counts=True)[1])
    print("Test labels unique:", np.unique(y_test), "counts:", np.unique(y_test, return_counts=True)[1])
except NameError:
    print("y_train/y_test not found yet - will perform split now.")

# --- Robust splitting: ensure both classes in train and test ---
# We'll try multiple random states until both classes appear in y_train (and optionally in y_test)
def robust_split(X, y, test_size=0.25, attempts=20):
    for rs in range(0, attempts):
        X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=test_size, random_state=rs, stratify=y)
        if len(np.unique(y_tr)) > 1 and len(np.unique(y_te)) > 0:
            return X_tr, X_te, y_tr, y_te, rs
    # fallback: no stratify
    X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=test_size, random_state=42, stratify=None)
    return X_tr, X_te, y_tr, y_te, None

X_train, X_test, y_train, y_test, used_rs = robust_split(X, y, test_size=0.25, attempts=50)
print("Used random_state:", used_rs)
print("y_train unique & counts:", np.unique(y_train, return_counts=True))
print("y_test unique & counts:", np.unique(y_test, return_counts=True))

# --- Retrain model (RandomForest) ---
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=350, max_depth=12, random_state=42)
rf.fit(X_train, y_train)

# check classes learned
print("rf.classes_:", rf.classes_)

# --- Safe predict_proba handling ---
proba = rf.predict_proba(X_test)
print("predict_proba shape:", proba.shape)

# If proba has 2 columns, take column for positive class (label==1)
if proba.shape[1] == 2:
    y_prob = proba[:, 1]
else:
    # only one column: map it to the class it represents
    # rf.classes_[0] is the single class RF learned; if it's 1 then probability=1 for positive class
    learned_class = rf.classes_[0]
    print("Model learned only one class:", learned_class)
    # If the model learned only class 0, then all probs for positive class are zeros
    if learned_class == 0:
        y_prob = np.zeros(len(proba))
    else:
        # learned_class == 1 => model gives probability 1 for positive class
        y_prob = np.ones(len(proba))

# --- Evaluate quickly (ROC needs both classes in y_test; if y_test has single class, ROC can't be computed) ---
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc, classification_report
y_pred = rf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred, zero_division=0))

# Compute ROC/AUC only if y_test has both classes
if len(np.unique(y_test)) > 1:
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    print("AUC:", roc_auc)
else:
    print("ROC/AUC unavailable: y_test has only one class:", np.unique(y_test))

```

```python
# === RESET & REBUILD: Reload original CSV and recreate binary labels safely ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc

# 1) Load original CSV (use the CSV inside dataset1 that we extracted earlier)
orig_path = "dataset1/kidney_disease.csv"
df_orig = pd.read_csv(orig_path)

# 2) Basic cleaning like before
df_orig.replace("?", np.nan, inplace=True)

# numeric columns (same as before)
numeric_cols = ["age","bp","sg","al","su","bgr","bu","sc","sod","pot","hemo","pcv","wc","rc"]

for col in numeric_cols:
    if col in df_orig.columns:
        df_orig[col] = pd.to_numeric(df_orig[col], errors='coerce')

# impute numeric medians
num_present = [c for c in numeric_cols if c in df_orig.columns]
df_orig[num_present] = df_orig[num_present].fillna(df_orig[num_present].median())

# fill categorical with mode
for c in df_orig.columns:
    if df_orig[c].dtype == "object":
        if not df_orig[c].mode().empty:
            df_orig[c] = df_orig[c].fillna(df_orig[c].mode().iloc[0])
        else:
            df_orig[c] = df_orig[c].fillna("unknown")

# 3) Check original target values BEFORE encoding/mapping
print("Original unique target values (raw):", df_orig['classification'].unique())
print("Original target counts (raw):\n", df_orig['classification'].value_counts(dropna=False))

# 4) Map target to binary correctly
# The raw dataset uses 'ckd' and 'notckd' or similar - handle both raw string and numeric encodings
# If values are strings, map 'notckd'->0 and 'ckd'->1. If numeric labels present (0,1,2) map accordingly.
# We will handle all possibilities robustly:

# If classification column is string-like, map directly
if df_orig['classification'].dtype == object:
    # normalize strings
    df_orig['classification'] = df_orig['classification'].str.strip().str.lower()
    # map common labels
    mapping = {}
    if 'notckd' in df_orig['classification'].unique():
        mapping.update({'notckd': 0})
    if 'ckd' in df_orig['classification'].unique():
        mapping.update({'ckd': 1})
    # apply mapping if found
    if mapping:
        df_orig['classification'] = df_orig['classification'].replace(mapping)
    else:
        print("Warning: string labels not recognized. Values:", df_orig['classification'].unique())

# If classification is numeric (0,1,2), map as we decided: 0->0 (normal), 2->1 (decline), drop tiny class 1
if df_orig['classification'].dtype != object:
    # show counts first
    print("Numeric label counts before mapping:\n", df_orig['classification'].value_counts())
    # drop label '1' if it exists and is just noise
    if 1 in df_orig['classification'].unique():
        # show how many samples are label 1
        c1_count = (df_orig['classification'] == 1).sum()
        print("Label '1' count:", c1_count)
        if c1_count <= 5:
            print("Dropping small/noise label '1' with", c1_count, "samples.")
            df_orig = df_orig[df_orig['classification'] != 1]
        else:
            print("Label '1' has >5 samples; not dropping automatically. Please inspect.")
    # map 0->0, 2->1 if 2 exists
    if 2 in df_orig['classification'].unique():
        df_orig['classification'] = df_orig['classification'].replace({0:0, 2:1})

# Final check of target
print("Final target unique:", df_orig['classification'].unique())
print("Final target counts:\n", df_orig['classification'].value_counts())

# 5) Encode categorical features (LabelEncoder) but avoid encoding numeric columns accidentally
from sklearn.preprocessing import LabelEncoder
df = df_orig.copy()
for c in df.columns:
    if df[c].dtype == object:
        df[c] = df[c].astype(str)  # ensure string
        df[c] = LabelEncoder().fit_transform(df[c])

# 6) Drop 'id' or any identifier columns if present
for col in ['id', 'ID', 'Id']:
    if col in df.columns:
        df = df.drop(columns=[col])
        print(f"Dropped identifier column: {col}")

# 7) Prepare X, y and verify class balance
X = df.drop("classification", axis=1)
y = df["classification"]
print("X shape:", X.shape, "y shape:", y.shape)
print("Class distribution:", y.value_counts())

# 8) Stratified split (guarantees both classes in train and test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
print("After split - train class counts:", y_train.value_counts(), "test class counts:", y_test.value_counts())

# 9) Train RandomForest
rf = RandomForestClassifier(n_estimators=300, max_depth=12, random_state=42)
rf.fit(X_train, y_train)
print("rf.classes_:", rf.classes_)

# 10) Safe predict_proba
proba = rf.predict_proba(X_test)
print("predict_proba shape:", proba.shape)
if proba.shape[1] == 2:
    y_prob = proba[:,1]
else:
    # fallback - construct y_prob based on learned class
    learned = rf.classes_
    print("Warning: RF returned single-column probabilities for classes:", learned)
    # if only class 0 learned -> prob positive=0, if only class1 -> prob positive=1
    if learned.shape[0] == 1 and learned[0] == 0:
        y_prob = np.zeros(len(proba))
    elif learned.shape[0] == 1 and learned[0] == 1:
        y_prob = np.ones(len(proba))
    else:
        raise RuntimeError("Unexpected predict_proba shape.")

# 11) Evaluate & show metrics + confusion matrix + AUC (if possible)
y_pred = rf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred, zero_division=0))

if len(np.unique(y_test)) > 1:
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    print("AUC:", auc(fpr, tpr))
else:
    print("AUC unavailable: test set has single class:", np.unique(y_test))
```

```python
# Robust SHAP diagnostic + fallback (run in NEW cell)
!pip install shap --quiet
import shap, numpy as np, pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.inspection import permutation_importance

out = Path(".")

# ensure X_test_df exists as a DataFrame with proper columns
try:
    X_test_df = X_test.copy()
    if not hasattr(X_test_df, "columns"):
        X_test_df = pd.DataFrame(X_test_df, columns=X.columns)
except Exception:
    X_test_df = pd.DataFrame(X_test, columns=X.columns)

try:
    X_train_df = X_train.copy()
    if not hasattr(X_train_df, "columns"):
        X_train_df = pd.DataFrame(X_train_df, columns=X.columns)
except Exception:
    X_train_df = pd.DataFrame(X_train, columns=X.columns)

print("X_train shape:", X_train_df.shape, "X_test shape:", X_test_df.shape)
print("Columns:", list(X.columns))

# 1) Try TreeExplainer.shap_values and print diagnostic info
explainer = shap.TreeExplainer(rf)
try:
    raw = explainer.shap_values(X_test_df)
    print("Type of raw:", type(raw))
    # show structures for list/ndarray/Explanation
    if isinstance(raw, list):
        print("len(raw) list:", len(raw))
        for i, r in enumerate(raw):
            print(f" raw[{i}] type:", type(r), "shape:", getattr(r, 'shape', None))
    elif isinstance(raw, np.ndarray):
        print("raw ndarray shape:", raw.shape)
    else:
        # shap.Explanation
        try:
            print("raw object has .values shape:", np.array(raw.values).shape)
        except Exception:
            print("raw object unknown structure")
except Exception as e:
    print("TreeExplainer.shap_values failed:", type(e), e)

# 2) Try the newer API: shap.Explainer with model.predict_proba (positive class)
try:
    # Explainer that explains probability of positive class
    expl = shap.Explainer(lambda x: rf.predict_proba(x)[:,1], X_train_df, feature_names=X.columns)
    shap_res = expl(X_test_df)   # This returns an Explanation object
    print("shap_res.values shape:", np.array(shap_res.values).shape)
    shap_vals = np.array(shap_res.values)  # shape (n_samples, n_features)
    # summary plot
    plt.figure(figsize=(8,6))
    shap.summary_plot(shap_vals, X_test_df, show=False)
    plt.tight_layout()
    plt.savefig(out/"shap_summary_newapi.png", dpi=200)
    plt.close()
    print("Saved shap_summary_newapi.png")
    shap_ok = True
except Exception as e:
    print("shap.Explainer approach failed:", type(e), e)
    shap_ok = False

# 3) If SHAP succeeded above we are done. If not, use permutation importance as fallback
if not shap_ok:
    print("SHAP failed  using permutation importance fallback.")
    r = permutation_importance(rf, X_test_df, y_test, n_repeats=20, random_state=42, n_jobs=-1)
    perm_imp = pd.Series(r.importances_mean, index=X.columns).sort_values(ascending=False)
    print("Top permutation importances:\n", perm_imp.head(10))
    plt.figure(figsize=(6,4))
    perm_imp.head(10).plot(kind='barh')
    plt.gca().invert_yaxis()
    plt.title("Permutation Importance (fallback)")
    plt.tight_layout()
    plt.savefig(out/"perm_importance.png", dpi=200)
    plt.close()
    print("Saved perm_importance.png")

# 4) Also save a simple dependence-like scatter for top feature (works for SHAP or fallback)
top_feat = None
if 'perm_imp' in locals():
    top_feat = perm_imp.index[0]
else:
    top_feat = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False).index[0]

plt.figure(figsize=(6,4))
plt.scatter(X_test_df[top_feat], rf.predict_proba(X_test_df)[:,1], alpha=0.6)
plt.xlabel(top_feat); plt.ylabel("Predicted probability (positive)")
plt.title(f"Dependence (approx) for {top_feat}")
plt.tight_layout()
plt.savefig(out/f"dependence_{top_feat}.png", dpi=200)
plt.close()
print("Saved dependence plot for", top_feat)

# 5) List created files
print("Saved files:", [f for f in out.iterdir() if f.suffix in ['.png', '.pkl']])
```
