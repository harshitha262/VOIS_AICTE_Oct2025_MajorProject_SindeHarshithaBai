**About Dataset**

**Context**

This project focuses on analyzing the Netflix dataset, which contains information about movies and TV shows available on the platform. The goal is to explore content trends, understand how Netflix’s catalog has grown over the years, and gain insights into factors like release patterns, countries, genres, and audience preferences.

**Data fields**

Show_Id- Unique ID for each show or movie
Category- Type of content (Movie or TV Show)
Title- Name of the movie or show
Director-Director of the content
Cast- Main actors involved
Country- Country of origin
Release_Date- Date the content was released
Rating- Content rating (e.g., PG, TV-MA)
Duration- Duration of the movie or number of seasons for shows
Type- Genre or category of content
Description- Short summary of the content

```python
def remove_outliers_column(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)
    return df

df = remove_outliers_column(df, 'Duration_Number')
df = remove_outliers_column(df, 'Title_Length')
df = remove_outliers_column(df, 'Cast_Count')
df = remove_outliers_column(df, 'Description_Words')
```
